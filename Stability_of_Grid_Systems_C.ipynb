# Preprocessing of Dataset Begins

```python
#dropping 'stab' feature because of it's direct relationship with the 'stabf' feature
df = df.drop(['stab'], axis = 1)
df.head(2)
```

```python
X =df.drop(['stabf'], axis = 1)
y =df['stabf']
```

```python
df.head()
```

```python
#Splitting the data into 80-20 train-test split 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#Using the standard scaler to transform the train set and X_test
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
normalised_X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
normallsed_X_test =  pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

```python
#importing evaluation metrics and classifiers
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
#Q2
TP = 255
FP = 1380
FN = 45
TN = 20
Precision = (TP/(TP+FP))
Recall =(TP/(TP+FN))
F1 = 2 * (Precision * Recall) / (Precision + Recall)
print(f'The F1_Score of this classifier is : {F1}')
```

```python
#Q14
#Creating a RandomForest Classifier
rfc = RandomForestClassifier(random_state = 1)
#Model Training
rfc.fit(normalised_X_train, y_train)
#the fit function trains the model with the training set inputs
```

```python
#prediciting on the test data
r_pred = rfc.predict(normallsed_X_test)
print('Accuracy: {}'.format(round(accuracy_score(y_test, r_pred),4)))
```

```python
#Q15
#Creating a XGboost Classifier
xgb = XGBClassifier(random_state = 1, learning_rate = 0.1, max_depth = 4)
#Model Training
xgb.fit(normalised_X_train, y_train)
```

```python
#prediciting on the test data
x_pred = rfc.predict(normallsed_X_test)
print('Accuracy: {}'.format(round(accuracy_score(y_test, x_pred), 4)))
```

```python
#Q16
#Creating a LGBM Classifier
lgbm = LGBMClassifier(random_state = 1)

#Model Training
lgbm.fit(normalised_X_train, y_train)

#prediciting on the test data
l_pred = lgbm.predict(normallsed_X_test)
print('Accuracy: {}'.format(round(accuracy_score(y_test, l_pred), 4)))
```

```python
#Q17
#Creating an ExtraTrees Classifier
extree = ExtraTreesClassifier(random_state = 1)

extree.fit(normalised_X_train, y_train)

ex_pred = extree.predict(normallsed_X_test)

print('Accuracy: {}'.format(accuracy_score(y_test, ex_pred)))
```

```python
#Q20
#creating a function to return the importance of each feature
def get_feature_importance(model, feat, col_name):
    ft_importance = pd.Series(model.feature_importances_, feat.columns).sort_values()
    importance_df = pd.DataFrame(ft_importance).reset_index()
    importance_df.columns = ['Features', col_name]
    importance_df[col_name].round(3)
    return importance_df
    
```

```python
feature_importance = get_feature_importance(extree, normalised_X_train, 'Feature_Importance')
feature_importance
```
